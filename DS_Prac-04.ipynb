```python
X = data.drop('MEDV', axis=1)
y = data['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=80, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load the trained model
model = LinearRegression()
model.fit(X_train, y_train)

# Define a function to predict the price
def predict_price(features):
    # Convert features to DataFrame
    new_house_df = pd.DataFrame(features)

    # Predict the price using the trained model
    predicted_price = model.predict(new_house_df)

    return predicted_price[0]

# Define the features dictionary
features = {
    'CRIM': [0.01],
    'ZN': [20.0],
    'INDUS': [8.0],
    'CHAS': [0.0],
    'NOX': [0.5],
    'RM': [6.0],
    'AGE': [70.0],
    'DIS': [4.0],
    'RAD': [4],
    'TAX': [300],
    'PTRATIO': [18.0],
    'B': [380.0],
    'LSTAT': [10.0]
}

# Predict the price using the defined features
predicted_price = predict_price(features)

# Display the predicted price
print("Predicted Price:", predicted_price)
```

```python
# Choose an index of a house from the dataset
house_index = 0  # For example, let's choose the first house

# Get the features of the chosen house from the dataset
house_features = X_test.iloc[house_index].values.reshape(1, -1)

# Get the actual price of the chosen house from the dataset
actual_price = y_test.iloc[house_index]

# Predict the price of the chosen house using the trained model
predicted_price = model.predict(house_features)

# Print the actual and predicted prices
print("Actual Price:", actual_price)
print("Predicted Price:", predicted_price[0])

# Calculate the difference between actual and predicted prices
difference = abs(actual_price - predicted_price[0])
print("Difference:", difference)
```
