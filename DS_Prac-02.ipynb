```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Student_prac02_dataset.csv")
data
```

```python
data_set = pd.DataFrame(data)
```

```python
data_set.isnull().sum()
```

```python
data_set.info()
```

```python
print(data_set.columns)
```

```python
unique_genders = data_set[' gender'].unique()
print("Unique Genders:", unique_genders)
```

```python
data_set[' gender'].value_counts()
```

```python
data_set.plot(kind='scatter', x=' test_score', y=' attendance', s=32, alpha=.8)
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([data_set[data_set[' gender'] == ' Male'][' cgpa'],
             data_set[data_set[' gender'] == ' Female'][' cgpa']],
            labels=['Male', 'Female'])
plt.title('Box Plot of CGPA by Gender')
plt.xlabel('Gender')
plt.ylabel('CGPA')
plt.show()
```

```python
sns.distplot(np.log(data_set[' test_score']))
```
