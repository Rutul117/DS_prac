# Write a code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.

# D. Matrix plot
## a) heat map

```python
df["sex"] = [1 if i == "male" else 0 for i in df["sex"]] 
df["embarked"] = [0 if i == "S" else i for i in df["embarked"]] 
df["embarked"] = [1 if i == "C" else i for i in df["embarked"]] 
df["embarked"] = [2 if i == "Q" else i for i in df["embarked"]] 
```

```python
cols = ["survived","pclass","sex","age","sibsp","parch","fare","embarked"]
```

```python
df_corr = df[cols].corr()
df_corr
```

```python
sbn.heatmap(df_corr,annot=True)
```

## Cluster map

```python
sbn.clustermap(df_corr,cmap='coolwarm',figsize=(6,5))
```
